```python
class Human:
    def __init__(self, height, weight):
        self.new_height = height
        self.new_weight = weight
        
    def walk(self):
        print("걸어다닙니다.")
        
class Athlete(Human):
    def __init__(self, height, weight, fat_rate) :
        super().__init__(height, weight)
        self.new_fat_rate = fat_rate
    
    def workout(self):
        print("운동을 합니다.")
        
h2 = Athlete(180, 85, 10)
h2.walk()
h2.workout()
```

```python
class TennisPlayer(Athlete):
    
    def workout(self):
        print("테니스를 합니다.")
        
h4 = TennisPlayer(180, 60, 10)
h4.walk()
h4.workout()
```

```python
class circle:
    def __init__(self, r):
        self.r = r;
    def get_area(self):
        area = self.r * self.r * 3.14
        return area
    def get_circumference(self):
        circumference = self.r * 2 * 3.14
        return circumference
        
radius = float(input("원의 반지름:"))
circle1 = circle(radius)

print("반지름 : %d" %radius)
print("원의 둘레 : %.2f" % circle1.get_circumference())
print("원의 면적 : %.2f" % circle1.get_area())
```

```python
try:
    print("나누기만 할 수 있는 계산기")
    nums = []
    nums.append(int(input("첫 번째 숫자 :")))
    nums.append(int(input("두 번째 숫자 :")))
    #nums.append(int(nums[0] / nums[1]))
    print("{0} / {1} = {2}".format(nums[0], nums[1], nums[2]))
except ValueError:
    print("에러! 잘못된 형식의 값 입력")
except ZeroDivisionError as err:
    print(err)
except Exception as err:
    print("알 수 없는 에러")
    print(err)
```

```python
try:
    print("한 자리 숫자만 나눌 수 있는 계산기")
    num1 = int(input("첫 번째 숫자를 입력하세요: "))
    num2 = int(input("두 번째 숫자를 입력하세요: "))
    if num1 >= 10 or num2 >= 10:
        raise ValueError
    print("{0} / {1} = {2}" .format(num1, num2, int(num1 / num2)))
except ValueError:
    print("잘못된 값을 입력, 한 자리 숫자만 입력")
except ZeroDivisionError:
    print("0으로 나누지 마세요")
else:
    print(num1 + num2)
```

```python
class BigNumberError(Exception):
    def __init__(self, msg):
        self.msg = msg
    def __str__(self):
        return self.msg
    
try:
    print("한 자리 숫자만 나눌 수 있는 계산기")
    num1 = int(input("첫 번째 숫자를 입력하세요: "))
    num2 = int(input("두 번째 숫자를 입력하세요: "))
    if num1 >= 10 or num2 >= 10:
        raise BigNumberError("입력값: {0}, {1}" .format(num1, num2))
    print("{0} / {1} = {2}" .format(num1, num2, int(num1 / num2)))
except ValueError:
    print("잘못된 값을 입력, 한 자리 숫자만 입력")
except BigNumberError as err:
    print("에러 발생, 한 자리 숫자만 입력")
    print(err)
finally:
    print("감사합니다")
```

```python
class Father:
    def __init__(self, father_name):
        self.father_name = father_name
    def print_father(self):
        print("부모 : " + self.father_name)
        
class Child(Father):
    def __init__(self, name1, name2):
        Father.__init__(self, name1)
        self.child_name = name2
    def print_child(self):
        print("아이 : " + self.child_name)
        
father1 = Father("김부모")
father1.print_father()
child1 = Child("박부모", "박자식")
child1.print_father()
child1.print_child()
```

```python
class TV(object):
    def __init__(self, size, year, company):
        self.size = size
        self.year = year
        self.company = company
    def describe(self):
        print(self.company + "에서 만든 " + self.year + "년형 " + self.size + "인치 TV")

class Laptop(TV):
    def describe(self):
        print(self.company + "에서 만든 " + self.year + "년형 " + self.size + "인치 노트북")
        
LG_TV = TV("32", "2021", "LG")
LG_TV.describe()

Samsung_Notebook = Laptop("15", "2020", "삼성")
Samsung_Notebook.describe()
```
